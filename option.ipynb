```python
strTemp = 'https://in.tradingview.com/chart/5eiBUcfm/?symbol=NSE%3A'
strter = 'https://web.sensibull.com/option-chain?expiry=2022-07-28&tradingsymbol='
```

```python
ls1 = df['Name']
ls2= df['company name']
ls3 = []
ls4 = []
```

```python
for i in range(len(ls1)):
    link = strTemp+ls1[i]
    link1 = strter+ls1[i]
    ls3.append(link)
    ls4.append(link1)
```

```python
import pyshorteners
shortener1 = pyshorteners.Shortener()
shortener2 = pyshorteners.Shortener()
z = 1
for i,j,k in zip(ls1,ls3,ls4):
    print(z,' trading view : ',shortener1.tinyurl.short(j),'|  sensibul :',shortener2.tinyurl.short(k),'| name :',i)
    z +=1
```
